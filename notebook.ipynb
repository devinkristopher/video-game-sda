````python
# Function to analyze platform performance by region
def top_platforms_by_region(df, region, top_n=5):
    sales_by_platform = (
        df.groupby('platform')[region]
        .sum()
        .sort_values(ascending=False)
        .head(top_n)
    )
    
    market_share = sales_by_platform / sales_by_platform.sum()
    
    result = pd.DataFrame({
        'total_sales_millions': sales_by_platform,
        'market_share': market_share
    })
    
    return result
````

# Step 5 : Hypothesis Tests

—Average user ratings of the Xbox One and PC platforms are the same.

—Average user ratings for the Action and Sports genres are different.

Set the *alpha* threshold value yourself.

Explain:

—How you formulated the null and alternative hypotheses

—What criteria you used to test the hypotheses~~,~~ and why


We will test two hypotheses: 

- ****That the average user ratings of the Xbox One and PC platforms are the same.****
- ****That the average ratings for the Action and Sports genres are different.****

In accordance with statistical convention, we will use a significance threshold of alpha = 0.05, or 5%.

****For the comparison of Xbox One and PC user ratings, the null hypothesis states that the average user ratings are equal; the alternative hypothesis states that they are different.****

****For the comparison of Action and Sports user ratings, the null hypothesis states that the average user ratings are equal; the alternative hypothesis states that they are different.****

We will use a t-test to evaluate these hypotheses. It will be based on a p-value; if the p-value is less than 0.05, the null hypothesis is rejected; otherwise, we cannot reject the hypothesis in question.

````python
xone_ratings = df_relevant[(df_relevant['user_score'] > 0) & (df_relevant['platform'] == 'XOne')]['user_score']
pc_ratings = df_relevant[(df_relevant['user_score'] > 0) & (df_relevant['platform'] == 'PC')]['user_score']

print('Sample size (Xbox One):', len(xone_ratings))
print('Sample size (PC):', len(pc_ratings))

alpha = 0.05  

results = st.ttest_ind(xone_ratings, pc_ratings)

print('p-value:', results.pvalue)
print('Mean user rating (Xbox One):', round(xone_ratings.mean(), 2))
print('Mean user rating (PC):', round(pc_ratings.mean(), 2))

if results.pvalue < alpha:
    print("Reject H0: The mean user ratings differ between Xbox One and PC")
else:
    print("Fail to reject H0: No statistically significant difference in mean user ratings between Xbox One and PC")
````
